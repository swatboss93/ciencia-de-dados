9. Coloque o **título do gráfico**, o **título do eixo-x** e o **título do eixo-y** para o gráfico scatter percapitas versus expectativas.

10. Mude as escala do eixo-x do gráfico anterior para "1k", "10k" e "100k".

```python
plt.scatter(percapita['valores'], expectativas['valores'])
plt.xscale('log')
plt.title("titulo do grafico")
plt.xlabel("titulo do eixo-x")
plt.ylabel('título do eixo-y')
plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])
```

11. Mude o tamanho das "círculos" do gráfico scatter com base na população armazenada no dataframe **populacoes**.

```python
plt.scatter(percapita['valores'], expectativas['valores'], s = populacoes['valores'])
plt.xscale('log')
plt.title("titulo do grafico")
plt.xlabel("titulo do eixo-x")
plt.ylabel('título do eixo-y')
plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])
```

12. Mude as cores das "círculos" do gráfico scatter com base no dataframe **cores**.

```python
plt.scatter(percapita['valores'], expectativas['valores'], s = populacoes['valores'], c = cores['valores'], alpha=.8)
plt.xscale('log')
plt.title("titulo do grafico")
plt.xlabel("titulo do eixo-x")
plt.ylabel('título do eixo-y')
plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])
```

13. Coloque o nome dos paíse "Índia"e "China" próximo aos dois círulos maiores do gráfico scatter. Além disto, ative o grid do gráfico.

```python
plt.scatter(percapita['valores'], expectativas['valores'], s = populacoes['valores'], c = cores['valores'])
plt.xscale('log')
plt.title("titulo do grafico")
plt.xlabel("titulo do eixo-x")
plt.ylabel('título do eixo-y')
plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])
plt.text(2300, 64.5, "Índia")
plt.text(4550, 72.7, "China")
plt.grid(True)
```

17. Mostre em um gráfico de pizza a porecentagem de sobreviventes por sexo armazenados do dataframe **df_titanic**.

```python
soma_sobreviventes = df_titanic['Survived'].groupby(df_titanic['Sex']).sum()

plt.pie(soma_sobreviventes, labels=soma_sobreviventes.index, autopct='%.2f')
```

```python
mortos = df_titanic['Survived'] == 0

df_nao_sobreviventes = df_titanic[mortos]

nao_sobreviventes = df_nao_sobreviventes['Survived'].groupby(df_nao_sobreviventes['Sex']).count()


plt.pie(nao_sobreviventes, labels=nao_sobreviventes.index, autopct='%.2f')
```

19. Mostre a quantidade de não sobreviventes por classe econômica (Pclass).


```python
nao_sobreviventes = df_nao_sobreviventes['Survived'].groupby(df_nao_sobreviventes['Pclass']).count()

plt.pie(nao_sobreviventes, labels=nao_sobreviventes.index, autopct='%.2f')
```
