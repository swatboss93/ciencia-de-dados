6. (Desafio) Escreva um programa em Python que apresente a frequências das palavras no arquivo **moby_dick.txt**. Dica: pesquise os comandos **sub** e **split**.

```python
import re
with open(arquivo, mode="r") as file:
    texto = file.read()
    
    texto_limpo = re.sub(r'[^\w\s]', '', texto)
    
    palavras = texto_limpo.lower().split()
    
    frequencias = {}
    
    for palavra in palavras:
        chave = palavra.lower()
        if chave in frequencias.keys():
            frequencias[chave] += 1
        else:
            frequencias[chave] = 1
    
    for palavra, frequencia in frequencias.items():
        print("{}: {}".format(palavra, frequencia))
    
```
